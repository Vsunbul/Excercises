## rock/ paper/ scissors game

```python

import random 

done = False

wins, loses, ties = 0, 0, 0

names = {"R": "Rock", "P": "Paper", "S": "Scissors"}

while not done:
    choice = input("Please Choose Next Move (R, S, P) (Q To Quit)").upper()
    cpu_choice = random.choice(["R", "P", "S"])
    if choice == cpu_choice:
        print(f"Its a tie! You both chose{names[choice]}")
        ties += 1
        
    elif choice == "R":
        if cpu_choice == "P":
            print(f"""CPU wins! You chose {names[choice]}, 
                  The CPU chose {names[cpu_choice]}""")
            loses += 1
            
        else:
            print(f"""You win! You chose {names[choice]}, 
                  The CPU chose {names[cpu_choice]}""")
            wins += 1
            
    elif choice == "P":
        if cpu_choice == "S":
            print(f"""CPU wins! You chose {names[choice]}, 
                  The CPU chose {names[cpu_choice]}""")
            loses += 1
            
        else:
            print(f"""You win! You chose {names[choice]}, 
                  The CPU chose {names[cpu_choice]}""")
            wins += 1
            
    elif choice == "S":
        if cpu_choice == "R":
            print(f"""CPU wins! You chose {names[choice]}, 
                  The CPU chose {names[cpu_choice]}""")
            loses += 1
            
        else:
            print(f"""You win! You chose {names[choice]}, 
                  The CPU chose {names[cpu_choice]}""")
            wins += 1
            
    elif choice == "Q":
        done = True
        
    else:
        print("Invalid Command")
    
print(f"current Stats: {wins} Wins, {loses} Loses, {ties} Ties")
```

```python

import random 

def write_to_txt(player_name:str, wins:int, loses:int, ties:int) -> int:
    """
    Appends player statistics to the "score.txt" file.
    """
    with open("score.txt", "a") as file:
        file.write(f"name : {player_name}  current Stats: {wins} Wins, {loses} Loses, {ties} Ties\n")
    file.close()

wins = 0
loses = 0
ties = 0
invalid = 0

names = {"R": "Rock", "P": "Paper", "S": "Scissors"}
lose_state = {"R": "P", "P": "S", "S": "R"}

while True:
       
    player_name = input("Please enter a player name : ").upper()
    if player_name == "Q":
        break
        
    wins = 0
    loses = 0
    ties = 0
    invalid = 0
    
    while wins + ties + loses < 3:
        choice = input("Please Choose Next Move (R, S, P) (Q To Quit)").upper()
        cpu_choice = random.choice(["R", "P", "S"])
        if choice == cpu_choice:
            print(f"Its a tie! You both chose{names[choice]}")
            ties += 1

        elif choice in ["R", "P", "S"]:
            if cpu_choice == lose_state[choice]:
                print(f"""CPU wins! You chose {names[choice]}, The CPU chose {names[cpu_choice]}""")
                loses += 1
                
            else:
                print(f"""You win! You chose {names[choice]}, The CPU chose {names[cpu_choice]}""")
                wins += 1

        elif choice == "Q":
            break
        
        else:
            print("Invalid Command")
            invalid += 1
            if invalid == 3:
                print("Invalid Command Please Try Again Later")
                break
            
  
    write_to_txt(player_name, wins, loses, ties)
    
    print(f"name : {player_name} current Stats: {wins} Wins, {loses} Loses, {ties} Ties")
```
