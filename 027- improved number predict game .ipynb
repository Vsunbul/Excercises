```python
print("""

A number has been chosen. This is stored in 
a variable called secret_number. 
Everyone running the program you will code is asked to play 
the Guess the Hidden Number game and 
guess which number is chosen for them. 
Those who don't guess the number will be stuck 
in an endless loop forever.

Your task is to write the program that implements
the above scenario.

The user will be asked to enter an integer;
While loop will be used.
It will be checked whether the number entered by the user
is the same as the Secret Number.
If the number the user selects is different from the Secret Number,
the user will say "Ha ha! You're stuck in my loop!" message and
the user will be asked to enter a number again.
If the number entered by the user matches the Secret Number, 
the number will be printed on the screen and 
the program will output: 
"Well done! You are free now."
""")

```

```python

def distance(guess_number:int):
    """
    This function allows estimating the distance of the entered 
    integer number to the secret number and guiding the user.
   
    input--->integer
    output--->string
    
    """

    if abs(secret_number - guess_number) > 100:
        print(""" Your guess is far from the secret number! \n 
             Ha ha! You're stuck in my loop!""")
            
    elif 10 < abs(secret_number - guess_number) < 50:
        print("""Your guess is close to the secret number, but not by much! \n 
             Number difference less than 50""")
            
    elif 5 < abs(secret_number - guess_number) < 10:
        print("""Your guess is very close to the secret number \n 
            Number difference less than 10""")
        
    elif 1 < abs(secret_number - guess_number) < 5:
        print("""Your guess is very close to the secret number \n 
            Number difference less than 5""")    
        
    elif secret_number == guess_number:
        print(" Well done! You are free now.")    


print(
"""
+==================================+
| WELCOME TO MY GAME!!!            |
| Please enter a integer value.    |
| which number I chose guess!!!    |
|                                  |
| So what is the secret number?    |
+==================================+
""")

secret_number = 777

while True:
    guess_number = input("Enter the guess number: ")
    
    if guess_number.isdigit() != True:
        print("Please enter digit type value.")
        continue
    guess_number = int(guess_number)
    
    if secret_number > guess_number:
        print("Secret number bigger than a guess number.")
        
        distance(guess_number)
        
    elif secret_number < guess_number:
        print("Guess number bigger than the secret number.") 
        distance(guess_number)
        
    else:
        print(" Well done! You are free now.")
        break
        

```
