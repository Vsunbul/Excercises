```python
try:
    f = open("fractions.txt", "r")
    nominator = [] 
    denominator = []
    
    for line in f:
        print(line.strip()) # line = "47 -10"
        nominator.append(int(line.strip().split("-")[0]))
        denominator.append(int(line.strip().split("-")[1]))
        #if one of the values is missing
    f.close()
    print(nominator)
    print(denominator)
                         
    for i in range(len(nominator)):
        
        #If there is an impossible operation
        try:               
            print(nominator[i], "/", denominator[i], "=", nominator[i] / denominator[i])
                 
        except ZeroDivisionError as err:
            print(nominator[i], "/", denominator[i], "=" " impossible")
                         
except FileNotFoundError as err:
    print(err)
#if one of the values is missing
except IndexError as  err:
    print(err)
#if there is str type data in the file
except ValueError as err:
    print(err)

    
```
