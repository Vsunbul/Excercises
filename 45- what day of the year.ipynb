```python
def leap_year(year:int)-> str:
    
   """
   This function checks if a year is a leap year.
   input: year -> int
   output: boolean -> T/F   
   """ 
   if year % 100 == 0:
        if year % 400 == 0:
            return True
        
        else:
            return False
    
   elif year % 4 == 0 :
        return True
        
   else:
        return False 
    
def days_in_month(year: int, months:int) -> int:
    """
       This function accurately generates the days of the months and
       outputs the day of the desired month.
   
       input : year -> integer 
             : months -> integer
     output: month -> integer
     """ 
    
    month_list = []
    for month in range(12):
        
        if month < 7:
            if month % 2 == 0:
                month_list.append(31)
            elif month == 1:
                if leap_year(year) == True:
                    month_list.append(29)
                else: 
                    month_list.append(28)
       
            elif month % 2 == 1:
                month_list.append(30)
    
        elif month >= 7:
            if month % 2 == 1:
                month_list.append(31)
           
            elif month % 2 == 0:
                month_list.append(30)
    return month_list[months]

def day_of_year(year, month, day):
    sum_days = 0
   
    for i in range(month):
        sum_days += days_in_month(year, i)
        
    return sum_days + day
    
    year = input("Enter a year : ")
    month = input("Enter a month : " )
    day = input("Enter a day : ")

                
while True: 
    year = input("Enter a year : ")
    month = input("Enter a month : " )
    day = input("Enter a day : ")

    if year.isdigit() != True or month.isdigit() != True or day.isdigit() != True:
        print("enter a digits value for date")
        
    year = int(year)
    month = int(month)
    day = int(day)
        
    if year < 1500:
        print("please enter a year bigger than 1500")
    
    if month < 0 or month > 12:
        print("Month must be between 1 and 12")
    
    if  day > days_in_month(year, month-1):
        print(" The number of days entered cannot exceed the number of that month.")
    
    else:
        if day_of_year(year, month-1, day) > 365:
            print("year: ", year + 1," total day:", 
                  day_of_year(year, month-1, day) - 365)
            break
        else:
            print("year: ", year, "month: ", month,"day: ",day, " total day:", 
                  day_of_year(year, month-1, day))
            break
```
