```python


import random

def number_generator(card_type:str) -> list:
    
    """
    
    This function generates a 15 or 16 digit 
    credit card number with the Luhn algorithm.
    
    input: card_type -> str
    output: card_number_list -> int
    
    """
    
    
    def calculations(card_type:str) -> int: 
        """
        
        This function makes the x * 2 operation
        in the Luhn algorithm independent of the 
        number length.
        input: card_type -> str
        output: card_number_list -> int
        
        """ 
       
        operated = []
        card_number_list = []
        card_type = int(card_type)
        
        for  i in range(card_type - 1):
            card_number_list.append(random.randint(0,9))
        
        #This process makes index jump in card numbers.
        for i in range(len(card_number_list)):
            sum_card_number = (card_number_list[i] * pow(2, ((i + (len(card_number_list) + 1) % 5) %2)))
            operated.append(sum_card_number) 
            
        control_number = 10 - (sum(operated) % 10)
        card_number_list.append(control_number)
        
        print("Generated card number:",*card_number_list)
    
    print("""
    --------------------------------
    Card Type
    Regular Credit Card Number : 16
    American Express Card Number: 15 
    --------------------------------
    """)
    
    if card_type == "16":
        print("Regular Card Type Selected")
        calculations(card_type)  
        
    elif card_type == "15":
        print("American Express Card Type Selected")
        calculations(card_type)
        
#Nentering card type operation.   
while True:
    card_type = input("please enter card number lenght. 15 or 16 :")
    if card_type.isdigit() != True:
        print("please enter 15 or 16")
        continue
 
    elif card_type == "15" or card_type == "16":
        number_generator(card_type)
        break
        
    else:
         print("please enter 15 or 16")
    
```
